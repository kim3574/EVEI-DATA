```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data, skipping the first 15 rows
data = pd.read_csv('data.csv', skiprows=15)

# Convert 'Time' to numeric, coercing errors to NaN
data['Time'] = pd.to_numeric(data['Time'], errors='coerce')

# Remove rows with NaN in 'Time' column (non-numeric values are now NaN)
data = data.dropna(subset=['Time'])

# Now that the problematic rows are removed, 'Time' is already in float format
# Determine the maximum lap number
max_lap = int(data['Lap'].max())

# Loop through each lap
for i in range(2, 7,2):
    lap = data[data['Lap'] == i]
    time = lap['Time']
    speed = lap['Speed (MPH)']
    
    # Subtract the first time value of the lap to start time from zero
    plt.plot(time - time.iloc[0], speed, label=f'Lap {i}')
plt.xlabel('Time (s)')
plt.ylabel('Speed (MPH)')
plt.legend()
plt.show()
```

```python
import numpy as np


lap_arr = []
time_arr = []
peak_arr = []

for i in range( max_lap + 1):
    lap_arr.append(i)
    lap = data[data['Lap'] == i]  # Filter data for each lap
    max(data[])
    time = lap['Time']
    end   = time.iloc[-1]
    start = time.iloc[0]
    diff = round(float(end) - float(start),2)
    time_arr.append(diff)
    

    
plt.bar(lap_arr, time_arr)

median_time = np.median(time_arr)
average_time = np.mean(time_arr)

plt.axhline(median_time, color='r', linestyle='-', label=f'Median: {median_time:.2f}')
plt.axhline(average_time, color='g', linestyle='-', label=f'Average: {average_time:.2f}')

for i, v in enumerate(time_arr):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom')

plt.xlabel('Lap')
plt.ylabel('Time Difference (Seconds)')
plt.title('Time Difference per Lap')
plt.legend()
plt.show()
```

```python
# Convert 'Time' column to timedelta
lap['Time'] = pd.to_timedelta(lap['Time'])

# Filter for lap 2
lap_2 = lap[lap['Lap'] == 2]

# Get the 'Time' and 'Speed (MPH)' columns
time = lap_2['Time']
speed = lap_2['Speed (MPH)']

# Get the start time of the lap and subtract it from all timestamps to start from 0
start_time = time.iloc[0]
time -= start_time

# Plotting
plt.plot(time, speed, label='Lap 2')

# Setting up x-ticks
n = 5
x_ticks = time.iloc[::len(time)//n]  # Ensure that x_ticks are selected at regular intervals
plt.xticks(x_ticks.dt.total_seconds(), x_ticks.astype(str))  # Convert timedelta to total seconds for plotting

plt.xlabel('Time')
plt.ylabel('Speed (MPH)')
plt.legend()
plt.show()
```
